```python
import numpy as np
from numpy import genfromtxt
```

```python
#exercise 2

shoppingChallenge = genfromtxt('DS1_C4_S1_Shopping_Data_Challenge.csv', delimiter=',', skip_header=1)
shoppingChallenge
```

```python
#task1 Display the details of the customers whose spending score is greater than 80. 
#Also, display the total count of these customers.

old=np.where(shoppingChallenge[:,3]>80)#filters the specific column index and returns the index of where those values are located
print(type(old))
arrayOld = np.array(old)

print(f'There are {arrayOld.size} customers who have a spending score higher than 80')

shoppingChallenge[old]#returns the actual values of my data
```

```python
#task2 Extract and display the data of the customers who are 20 to 24 years of age.

age24 = np.where((shoppingChallenge[:,1] >= 20) & (shoppingChallenge[:,1]<=24))
shoppingChallenge[age24]
```

```python
#task3 Display the IDs of the customers with the lowest annual income and the customers with the highest.

highest_income = np.sort(shoppingChallenge)[::-1]
print(f"The ID's of the top 10 highest incomes are: \n {highest_income[0:11]} ")
print("")
lowest_income = np.sort(shoppingChallenge)
print(f"The ID's of the top 10 highest incomes are: \n {lowest_income[0:11]} ")
```

```python
#task2.4 Display the details of the customers who are less than 45 years of age, 
#have an annual income greater than $35k, and have a spending score greater than 75. 
#How many such customers are there in the given data?

multi_condition = np.where((shoppingChallenge[:,1] < 45) & (shoppingChallenge[:,2] >35) & (shoppingChallenge[:,3] > 75))
shoppingChallenge[multi_condition]

print(f'There are {np.array(multi_condition).size} customers that are less than 45 years,make more than 35k, and have a spending score higher than 75')
shoppingChallenge[multi_condition]
```

```python
# task2.5- During the Christmas season, the company wishes to offer random gift coupons 
# valued between $10 and $50 to all its customers. Add a new column for the gift coupons in the existing data. 
# Display the details of the customers after adding the new column.

print(shoppingChallenge.shape)

randomDiscount = np.random.randint(10,51,200)#generates 200 random values between 10 and 50 and puts it into a 1D array

newShape = randomDiscount.reshape(200,1) #reshapes it from (200,_) into a 2D array (200,1)

#newColumn = np.hstack((shoppingChallenge,newShape))
newColumn = np.append(shoppingChallenge,newShape, axis = 1)#adds both arrays together or in other words, it adds a column

newColumn
```
